```python
import pandas as pd, numpy as np
df = pd.read_csv("viterbi_debug_top10.csv")  # nid,rank,global_index,id_trc,direction,score,backptr,backptr_id_trc

# only top-1 per nid (the state Viterbi actually picks for each nid)
top1 = df[df['rank']==1].copy()

# parity = direction (even=0 forward, odd=1 reverse)
parity_cur  = top1['global_index'] % 2
parity_prev = top1['backptr']      % 2

# how often the backpointer changes parity (likely flips direction)
flip_rate = (parity_cur != parity_prev).mean()
adjacent_rate = (abs(top1['global_index'] - top1['backptr']) == 1).mean()  # same edge, opposite direction

print("Parity flip rate (prev vs cur):", flip_rate)
print("Backptr is adjacent (±1) rate :", adjacent_rate)

## 95 % of states,
## That means in 95 % of states,
## the backpointer you follow during Viterbi backtrace jumps to the opposite-parity state —
## if the current state index is even (forward), the backpointer index is odd (reverse), and vice versa.
```
